```python

from sympy import *
from IPython.display import display, Math
init_printing()

def verify_steps(steps):
    """Sprawdź i wyświetl wszystkie kroki przekształceń"""
    
    for i in range(len(steps)):
        print("-" * 30)
        display(steps[i])
        
        # Weryfikuj tylko jeśli nie jest to pierwszy krok
        if i > 0:
            current = steps[i-1]
            new = steps[i]
            diff = simplify(current.lhs - current.rhs - (new.lhs - new.rhs))
            
            if diff == 0:
                print("✅")
            else:
                print("❌")
                display(diff)
        
        print()

# zmienne
y_hat, y_bar, b, x, x_bar = symbols('y_hat y_bar b x x_bar')

# kroki
steps = [
    Eq(y_hat - y_bar, b*(x - x_bar)),          # wzór początkowy
    Eq(y_hat - y_bar, b*x - b*x_bar),          # rozwijam nawias
    Eq(y_hat, y_bar + b*x - b*x_bar),          # przenoszę y_bar
    Eq(y_hat, y_bar + b*(x - x_bar))           # grupuję z powrotem
]

# weryfikacja
verify_steps(steps)
```

```python
import sympy as sp
from IPython.display import display, Markdown
import re

# Definicja symboli i funkcji SymPy
i, j, n = sp.symbols('i j n', integer=True, positive=True)
x_bar, y_bar, b_hat = sp.symbols('x_bar y_bar b_hat')
x = sp.Function('x')
y = sp.Function('y')
y_hat = sp.Function('y_hat')

# Funkcja do konwersji SymPy na ładny LaTeX
def pretty_latex(expr):
    latex_str = sp.latex(expr)
    # Zamień x_bar na \bar{x}
    latex_str = latex_str.replace('x_{bar}', r'\bar{x}')
    latex_str = latex_str.replace('y_{bar}', r'\bar{y}')
    latex_str = latex_str.replace('b_{hat}', r'\hat{b}')
    
    # Zamień y_{hat}{\left(i \right)} na \hat{y}_i
    latex_str = re.sub(r'y_{hat}\{\\left\(([^}]+) \\right\)\}', r'\\hat{y}_{\1}', latex_str)
    # Zamień x{\left(i \right)} na x_i
    latex_str = re.sub(r'x\{\\left\(([^}]+) \\right\)\}', r'x_{\1}', latex_str)
    # Zamień y{\left(i \right)} na y_i
    latex_str = re.sub(r'y\{\\left\(([^}]+) \\right\)\}', r'y_{\1}', latex_str)
    
    return latex_str

def display_expr(expr, text=""):
    if text:
        print(text)
    latex_code = pretty_latex(expr)
    display(Markdown(f"$${latex_code}$$"))

# Teraz możesz programować w SymPy i ładnie wyświetlać
left_side = sp.Sum(2*(y_hat(i) - y_bar)*(y(i) - y_hat(i)), (i, 1, n))
right_side_1 = 2*b_hat*sp.Sum((x(i) - x_bar)*(y(i) - y_hat(i)), (i, 1, n))
right_side_2 = 2*b_hat*sp.Sum((x(i) - x_bar)*((y(i) - y_bar) - b_hat*(x(i) - x_bar)), (i, 1, n))

# Wyświetlanie
display_expr(left_side, "Lewa strona:")
print("=")
display_expr(right_side_1)
print("=")
display_expr(right_side_2)
```
